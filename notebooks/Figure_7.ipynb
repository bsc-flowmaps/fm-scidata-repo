```python
from flowmaps_data import geolayer, covid19, population
import plotly.graph_objects as go
    
date = '2020-10-10'
layer = 'cnig_provincias' # 'abs_09'
ev = 'ES.covid_cpro' # '09.covid_abs'
    
# Download geojson layer
geojson = geolayer(layer)
    
# Download population
pop_df = population(layer=layer, start_date=date, end_date=date)
    
# Download COVID-19 data
covid_df = covid19(ev=ev, start_date=date, end_date=date)
    
# Combine COVID-19 cases and population
# combined_df = covid_df.merge(pop_df, left_on='id', right_on='id')
covid_df['cases_by_100k'] = 100000 * covid_df['new_cases'] / covid_df['population']
    
# Plot
fig = go.Figure(go.Choroplethmapbox(geojson=geojson,
                                        locations=covid_df['id'],
                                        z=covid_df['cases_by_100k'],
                                        colorscale="Reds",
                                        colorbar={'title': 'Cases by 100k<br>population'},
                                        marker_opacity=0.8))
fig.update_layout(title=f'Covid-19 daily incidence by 100k population at {date}',
                      mapbox_style="carto-positron",
                      mapbox_zoom=4.5,
                      mapbox_center={"lat": 40.495178477814555, "lon": -3.717336960173357})
fig.update_layout(margin={"r":0,"t":30,"l":0,"b":0})
fig.show()
```
